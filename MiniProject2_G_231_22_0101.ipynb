```python
plt.figure(figsize=(11,6))

bulan = np.arange(1,13)

plt.bar(bulan, df['1981 - 1990'], width = 0.2, color='blue', label='1981 - 1990')
plt.bar(bulan + 0.2, df['1991 - 2000'], width = 0.2, color='red', label='1991 - 2000')
plt.bar(bulan + 0.2 * 2, df['2001 - 2010'], width = 0.2, color='green', label='2001 - 2010')

for i in range(20, 700, 20):
  plt.axhline (y=i, color='grey', linestyle='-', linewidth=0.3, zorder=1)

for i in range(100, 700, 100):
  plt.axhline (y=i, color='black', linestyle='-', linewidth=0.2, zorder=1)

plt.ylim(0,700)

plt.xlabel('Bulan', fontsize=12, fontweight='bold')
plt.ylabel('Curah Hujan (mm)', fontsize=12, fontweight='bold')

plt.legend(loc='upper center', ncols=3, frameon = False, bbox_to_anchor=(0.5,1.1))

plt.xticks(bulan + 0.25, bulan)

plt.show()
```

```python
suhupermukaan = pd.read_csv('suhupermukaan.csv')
suhupermukaan
```

```python
waktu = suhupermukaan['Waktu']
asimilasi = suhupermukaan['Asimilasi']
observasi = suhupermukaan['Observasi']
non_asimilasi = suhupermukaan['Non Asimilasi']

plt.figure(figsize = (12,6))

plt.plot(waktu, asimilasi, color='blue', marker='o', linewidth=2.5, label='Asimilasi')
plt.plot(waktu, observasi, color='orange', marker='o', linewidth=2.5, label='Observasi')
plt.plot(waktu, non_asimilasi, color='red', marker='o', linewidth=2.5, label='Non Asimilasi')

for celcius in range(20,27):
  plt.axhline(y=celcius, color='grey',linestyle='-', linewidth=0.4)

plt.title('Suhu Permukaan', color= 'k', fontsize=14)

plt.xlabel('Waktu', color='k', fontsize=12,)
plt.ylabel('Celcius (°C)', color='k', fontsize=12)

plt.xticks(rotation=15)

plt.ylim(20,27)

plt.legend(loc='center right', frameon=False, bbox_to_anchor=(1.2, 0.5))

plt.show()
```
